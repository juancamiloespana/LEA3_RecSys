```python

import os  ### para ver y cambiar directorio de trabajo
import pandas as pd
import sqlite3 as sql ### para conectarse a BD


##### conectarse a BD #######
conn= sql.connect('data/db_movies')
cur=conn.cursor()

### para ver las tablas que hay en la base de datos
cur.execute("select name from sqlite_master where type='table' ")
cur.fetchall()
```

```python
####3
pd.read_sql("""select movieId, avg(rating)
            from ratings
            where movieId=1
            group by movieId
            order by userId asc""", conn)
```

```python

####4
pd.read_sql("""select a.title, count(b.rating) as cnt
            from movies a left join ratings b on a.movieId=b.movieId
            where b.rating is null
            group by a.title
           order by cnt asc """, conn)




```

```python
####5
pd.read_sql("""select a.title, count(b.rating) as cnt
            from movies a inner join ratings b on a.movieId=b.movieId
            group by a.title
            having cnt=1
            order by cnt asc """, conn)
```

```python
####6
pd.read_sql("""select genres, count(*) as cnt
            from movies
            group by genres
            order by cnt desc
            limit 1 offset 8""", conn)
```

```python
### alternativa para usar with ##

pd.read_sql("""with t1 as (select genres, count(*) as cnt
            from movies
            group by genres
            order by cnt desc limit 9),
            t2 as (select * from movies)
            select * from t1 order by cnt asc limit 1 """, conn)
```
